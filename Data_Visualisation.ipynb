# Data Visualization
Data visualization is the representation of data and information in graphical or visual formats to help understand patterns, trends, and insights that may be difficult to discern from raw data alone. It is a way to communicate complex ideas effectively and make data more accessible and meaningful.

## Coordinate Systems and Axes

Cartesian Coordinate System:
The Cartesian coordinate system, also known as rectangular coordinate system, is a common framework used in data visualization. It consists of two perpendicular axes, the x-axis (horizontal) and the y-axis (vertical), intersecting at a point called the origin (0, 0). Each axis has a scale that represents the numerical values of the data being plotted.

Axes:
Axes are the lines or line segments that define the x-axis and y-axis in a coordinate system. They provide a reference for positioning data points and other visual elements. The axes are typically labeled with tick marks and values to indicate the scale of the data.

Scale:
The scale of an axis determines how the data values are mapped to the coordinate system. It defines the range of values that can be displayed and the corresponding positions along the axis. Common types of scales include linear scales, logarithmic scales, and categorical scales.

Axis Labels:
Axis labels are textual descriptions that indicate the quantity or variable represented by each axis. They provide important context and help users interpret the visualization. For example, the x-axis might represent time, and the y-axis might represent temperature.

Axis Ticks:
Axis ticks are marks along an axis that indicate specific values. They help users identify the position and magnitude of data points. Major ticks are usually labeled with values, while minor ticks provide additional reference points.

Axis Titles:
Axis titles are descriptive labels placed near the axes that provide a brief explanation of the represented variable or quantity. They help users understand the purpose of each axis and the meaning of the data being visualized.

Coordinate Units:
Coordinate units define the scale or granularity of the data being represented. They can be continuous, such as real numbers or dates, or discrete, such as categories or labels. The appropriate choice of coordinate units depends on the nature of the data and the insights being conveyed.

Origin:
The origin is the point of intersection between the x-axis and y-axis in a Cartesian coordinate system. It has coordinates (0, 0) and serves as the reference point for positioning data points. The position of data points relative to the origin indicates their x and y values.

Titles:
Titles are used to provide a descriptive and concise summary of the visualization or the main message it conveys. They are typically placed at the top of the figure or chart.

Captions:
Captions are explanatory texts placed below the visualization to provide additional details, interpretations, or explanations of the data.

Tables:
Tables are used to present structured data in tabular format, providing a clear and organized way to display information.



```python
!pip install squarify
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Generate data

x = np.linspace(0, 5, 50)
x1 = np.linspace(0, 10, 50)
x2 = np.linspace(0, 15, 50)
x3 = np.linspace(0, 20, 50)
x4 = np.linspace(0, 25, 50)

y = np.sin(x)
y1 = np.sin(x1)
y2 = np.cos(x2)
y3 = np.sin(x3)
y4 = np.cos(x4)
```

### Create a figure and axes

```python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot the data
ax.plot(x1, y1, label='Sin(x)', color='blue', linestyle='--', linewidth=2)
ax.plot(x2, y2, label='Cos(x)', color='red', linestyle='-', linewidth=2)

# Set labels for x-axis and y-axis
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Set title
ax.set_title('Trigonometric Functions')

# Set axis limits
ax.set_xlim([0, 10])
ax.set_ylim([-1.2, 1.2])

# Add a grid
ax.grid(True, linestyle='--', alpha=0.5)

# Add legend
ax.legend()

# Add text annotation
ax.text(5, 1, 'Maximum', ha='center', va='bottom', color='green', fontsize=12)

# Add arrow annotation
ax.annotate('Inflection Point', xy=(np.pi/2, 0), xytext=(3, 0.7),
            arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)

# Add vertical and horizontal lines
ax.axvline(x=np.pi/2, color='gray', linestyle='--', linewidth=1)
ax.axhline(y=0, color='gray', linestyle='--', linewidth=1)

# Customize tick labels
ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
ax.set_xticklabels(['0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$'])
ax.set_yticks([-1, -0.5, 0, 0.5, 1])

# Save the plot
plt.savefig('plot.png', dpi=300)

# Show the plot
plt.show()
```

## Color Scales
Color scales play a crucial role in data visualization as they enable the representation of data values through color gradients.

### Sequential Color Scales
Sequential color scales are used to represent ordered data that has a natural progression or magnitude. They typically start with a light color for low values and transition to a darker color for high values. Examples include "Blues," "Greens," and "Oranges."

```python
# Sequential Color Scale

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the data using a sequential color scale
ax.plot(x, y, color='blue')

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Sequential Color Scale')

# Show the plot
plt.show()
```

### Diverging Color Scales
Diverging color scales are used to represent data with a central point or critical value. The colors diverge from a central color to two contrasting colors, highlighting both positive and negative deviations from the critical value. Examples include "RdBu" (Red-Blue), "PiYG" (Pink-Yellow-Green), and "PRGn" (Purple-Red-Green).


```python
# Diverging Color Scale
import matplotlib.cm as cm

# Generate data
x = np.linspace(-5, 5, 100)
y = np.tanh(x)

# Create a figure and axes
fig, ax = plt.subplots()

# Define a diverging colormap
cmap = cm.get_cmap('RdBu')

# Plot the data using a colormap
ax.plot(x, y, c=cmap(0.5))

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Diverging Color Scale')

# Show the plot
plt.show()
```

### Qualitative Color Scales
Qualitative color scales are used to represent categorical or unordered data without any inherent order or magnitude. They consist of distinct and distinguishable colors without any specific progression. Examples include "Set1," "Pastel1," and "Dark2."


```python
# Qualitative Color Scale
import matplotlib.cm as cm

# Generate data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Create a figure and axes
fig, ax = plt.subplots()

# Define a qualitative colormap
cmap = cm.get_cmap('Dark2')

# Determine the number of segments
num_segments = len(x) - 1

# Plot the line segments with different colors
for i in range(num_segments):
    color = cmap(i % cmap.N)
    ax.plot(x[i:i+2], y[i:i+2], color=color)

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Qualitative Color Scale')
```

### Cyclical Color Scales
Cyclical color scales are used to represent data that exhibits periodic or cyclic patterns. These scales wrap around a color wheel, providing a smooth transition from one color to another. Examples include "Hue," "Twilight," and "Rainbow."

```python
# Cyclical Color Scale

# Generate data
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Create a figure and axes
fig, ax = plt.subplots()

# Define a sequence of colors for the cyclical color scale
colors = ['purple', 'navy', 'green', 'gold', 'red']

# Plot the data with a cyclical color scale
for i in range(len(x)-1):
    ax.plot(x[i:i+2], y[i:i+2], color=colors[i % len(colors)])

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Cyclical Color Scale')

# Show the plot
plt.show()
```

#### Bar plots :
Bar plots are a popular type of data visualization that use rectangular bars to represent and compare different categories or groups.

```python
# Vertical Bar Chart

# Generate Data
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the vertical bar chart
ax.bar(categories, values, align='center')

# Set labels for x-axis and y-axis
ax.set_xlabel('Categories')
ax.set_ylabel('Values')

# Set title
ax.set_title('Vertical Bar Chart')

# Show the plot
plt.show()
```

```python
# Horizontal Bar Chart

# Generate Data
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the horizontal bar chart
ax.barh(categories, values)

# Set labels for x-axis and y-axis
ax.set_xlabel('Values')
ax.set_ylabel('Categories')

# Set title
ax.set_title('Horizontal Bar Chart')

# Show the plot
plt.show()
```

```python
# Grouped bar chart

# Generate Data
categories = ['A', 'B', 'C', 'D']
values1 = [10, 15, 7, 12]
values2 = [8, 12, 9, 10]

# Set the width of each bar
bar_width = 0.35

# Create a figure and axes
fig, ax = plt.subplots()

# Calculate the positions for the bars
bar_positions1 = np.arange(len(categories))
bar_positions2 = bar_positions1 + bar_width

# Plot the grouped bar chart
ax.bar(bar_positions1, values1, width=bar_width, label='Group 1')
ax.bar(bar_positions2, values2, width=bar_width, label='Group 2')

# Set labels for x-axis and y-axis
ax.set_xlabel('Categories')
ax.set_ylabel('Values')

# Set title
ax.set_title('Grouped Bar Chart')

# Set x-axis tick labels
ax.set_xticks(bar_positions1 + bar_width / 2)
ax.set_xticklabels(categories)

# Add legend
ax.legend()

# Show the plot
plt.show()
```

```python
# Stacked Bar Chart

# Generate Data
categories = ['A', 'B', 'C', 'D']
values1 = [10, 15, 7, 12]
values2 = [8, 12, 9, 10]

# Create a figure and axes
fig, ax = plt.subplots()

# Calculate the positions for the bars
bar_positions = np.arange(len(categories))

# Plot the stacked bar chart
ax.bar(bar_positions, values1, label='Group 1')
ax.bar(bar_positions, values2, bottom=values1, label='Group 2')

# Set labels for x-axis and y-axis
ax.set_xlabel('Categories')
ax.set_ylabel('Values')

# Set title
ax.set_title('Stacked Bar Chart')

# Set x-axis tick labels
ax.set_xticks(bar_positions)
ax.set_xticklabels(categories)

# Add legend
ax.legend()

# Show the plot
plt.show()
```

### Visualizing Distributions:

#### Histogram:
A histogram represents the distribution of a continuous variable by dividing the data into bins and showing the frequency or density of observations in each bin.

```python
# Histogram

# Generate random data
data = np.random.randn(1000)

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the histogram
ax.hist(data, bins=30, density=True)

# Set labels and title
ax.set_xlabel('Values')
ax.set_ylabel('Frequency')
ax.set_title('Histogram')

# Show the plot
plt.show()
```

#### Kernel Density Plot:
A kernel density plot provides a smooth estimate of the probability density function of a continuous variable. It visualizes the distribution as a smooth curve.

```python
# Kernel Density Plot

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the kernel density estimate
sns.kdeplot(data, ax=ax)

# Set labels and title
ax.set_xlabel('Values')
ax.set_ylabel('Density')
ax.set_title('Kernel Density Plot')

# Show the plot
plt.show()
```

#### Box Plot:
A box plot (or box-and-whisker plot) displays the summary statistics of a numerical variable, including the median, quartiles, and potential outliers. It gives an overview of the data distribution and identifies any skewness or outliers.

```python
# Box Plot

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the box plot
ax.boxplot(data)

# Set labels and title
ax.set_xlabel('Variable')
ax.set_ylabel('Values')
ax.set_title('Box Plot')

# Show the plot
plt.show()
```

#### Violin Plot:
A violin plot combines a box plot and a kernel density plot. It shows the distribution of a numerical variable, including the summary statistics and the density estimate as a rotated kernel density plot.

```python
# Violin Plot

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the violin plot
sns.violinplot(data, ax=ax)

# Set labels and title
ax.set_xlabel('Variable')
ax.set_ylabel('Values')
ax.set_title('Violin Plot')

# Show the plot
plt.show()
```

#### Rug Plot:
A rug plot is a one-dimensional scatter plot that displays individual data points as tick marks along an axis. It provides a visual representation of the data distribution.

```python
# Rug Plot

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the rug plot
ax.plot(data, np.zeros_like(data), '|')

# Set labels and title
ax.set_xlabel('Values')
ax.set_title('Rug Plot')

# Remove y-axis
ax.get_yaxis().set_visible(False)

# Show the plot
plt.show()
```

#### Empirical Cumulative Distribution Function (ECDF) Plot:
An ECDF plot displays the cumulative probability of observing a value less than or equal to a given data point. It helps understand the overall distribution and percentiles of a variable.

```python
# ECDF Plot

from statsmodels.distributions.empirical_distribution import ECDF

# Calculate the empirical cumulative distribution function
ecdf = ECDF(data)

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the ECDF
ax.plot(ecdf.x, ecdf.y)

# Set labels and title
ax.set_xlabel('Values')
ax.set_ylabel('Cumulative Probability')
ax.set_title('ECDF Plot')

# Show the plot
plt.show()
```

#### QQ Plot:
A QQ (quantile-quantile) plot compares the quantiles of a variable against the expected quantiles of a theoretical distribution, typically the normal distribution. It helps assess the similarity between the data distribution and the theoretical distribution.

```python
# QQ Plot

import scipy.stats as stats

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the QQ plot
stats.probplot(data, dist='norm', plot=ax)

# Set title
ax.set_title('QQ Plot')

# Show the plot
plt.show()
```

#### Probability Plot:
A probability plot visually assesses whether a data distribution follows a specific theoretical distribution. It compares the quantiles of the data against the quantiles of the theoretical distribution.

```python
# Probability Plot

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the probability plot
stats.probplot(data, dist='norm', plot=ax)

# Set title
ax.set_title('Probability Plot')

# Show the plot
plt.show()
```

### Visualizing Many Distributions at Once
When you need to visualize many distributions at once, you can use techniques like overlaying multiple histograms, creating violin plots, or using box plots with additional enhancements.

```python
# Overlaid Histograms

# Generate random data for multiple distributions
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(2, 0.5, 1000)
data3 = np.random.normal(-1, 2, 1000)

# Create a figure and axes
fig, ax = plt.subplots()

# Plot overlaid histograms
ax.hist(data1, bins=30, alpha=0.5, label='Distribution 1')
ax.hist(data2, bins=30, alpha=0.5, label='Distribution 2')
ax.hist(data3, bins=30, alpha=0.5, label='Distribution 3')

# Set labels and title
ax.set_xlabel('Values')
ax.set_ylabel('Frequency')
ax.set_title('Overlaid Histograms')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

```python
# Violin Plots

# Combine the data into a list
data = [data1, data2, data3]

# Create a figure and axes
fig, ax = plt.subplots()

# Plot violin plots
ax.violinplot(data, showmedians=True)

# Set labels and title
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Distribution 1', 'Distribution 2', 'Distribution 3'])
ax.set_ylabel('Values')
ax.set_title('Violin Plots')

# Show the plot
plt.show()
```

```python
# Box Plots

# Create a figure and axes
fig, ax = plt.subplots()

# Plot box plots
ax.boxplot(data, notch=True, sym='')

# Set labels and title
ax.set_xticks([1, 2, 3])
ax.set_xticklabels(['Distribution 1', 'Distribution 2', 'Distribution 3'])
ax.set_ylabel('Values')
ax.set_title('Box Plots')

# Show the plot
plt.show()
```

```python
# Strip Plot

# Create a figure and axes
fig, ax = plt.subplots()

# Plot strip plot
sns.stripplot(data=data, jitter=True, ax=ax)

# Set labels and title
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['Distribution 1', 'Distribution 2', 'Distribution 3'])
ax.set_ylabel('Values')
ax.set_title('Strip Plot')

# Show the plot
plt.show()
```

```python
# Swarm Plot

# Create a figure and axes
fig, ax = plt.subplots()

# Plot swarm plot
sns.swarmplot(data=data, ax=ax)

# Set labels and title
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(['Distribution 1', 'Distribution 2', 'Distribution 3'])
ax.set_ylabel('Values')
ax.set_title('Swarm Plot')

# Show the plot
plt.show()
```

```python
# KDE Plot

# Combine the data into a list
data = [data1, data2, data3]

# Create a figure and axes
fig, ax = plt.subplots()

# Plot KDE plot
for d in data:
    sns.kdeplot(d, ax=ax)

# Set labels and title
ax.set_xlabel('Values')
ax.set_ylabel('Density')
ax.set_title('KDE Plot')

# Show the plot
plt.show()
```

### Visualizing Proportions

### Visualizing Nested Proportions
Visualizing nested proportions is commonly done using a hierarchical chart known as a nested pie chart or a sunburst chart. This type of chart displays hierarchical data with nested circles or arcs, representing the proportions at different levels of the hierarchy.

```python
import matplotlib.pyplot as plt

# Generate Data
categories = ['Category A', 'Category B', 'Category C']
proportions = [0.6, 0.3, 0.1]
subcategories = ['Subcategory 1', 'Subcategory 2', 'Subcategory 3', 'Subcategory 4']
sub_proportions = [0.4, 0.3, 0.2, 0.1]

# Create a figure and axes

fig, ax = plt.subplots(figsize = ([12, 6]))

# Plot the outer bar chart
ax.bar(categories, proportions, color=['#ff9999', '#66b3ff', '#99ff99'])

# Plot the inner bar chart
inner_colors = ['#ffcccc', '#ccebff', '#ccffcc', '#e6ffb3']
for i in range(len(categories)):
    start = sum(proportions[:i])
    ax.bar(subcategories, [sub_proportions[i]*proportions[i]]*len(subcategories),
           bottom=[start]*len(subcategories), color=inner_colors[:len(subcategories)])

# Set labels and title
ax.set_xlabel('Categories')
ax.set_ylabel('Proportions')
ax.set_title('Nested Proportions')

# Show the plot
plt.show()
```

### Visualizing Time Series and Other Functions of an Independent Variable

#### Line Plot:
A line plot is commonly used to visualize time series data, where the independent variable represents time and the dependent variable represents the value at each time point.

```python
# Generate time series data
time = np.arange(0, 10, 0.1)
data = np.cos(time)

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the line plot
ax.plot(time, data)

# Set labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Value')
ax.set_title('Time Series')

# Show the plot
plt.show()
```

#### Area Plot:
An area plot can be used to visualize cumulative functions or stacked values of the dependent variable over time or the independent variable.

```python
# Generate time series data
time = np.arange(0, 10, 0.1)
data1 = np.sin(time)
data2 = np.cos(time)

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the area plot
ax.fill_between(time, data1, data2, alpha=0.5)

# Set labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Value')
ax.set_title('Area Plot')

# Show the plot
plt.show()
```

#### Scatter Plot with Regression Line:
A scatter plot with a regression line allows you to visualize the trend while also displaying individual data points. It provides a clear view of the relationship between variables and helps identify the direction and strength of the trend.

```python
# Generate time series data
x = np.arange(0, 10, 0.1)
y = 2 * x + np.random.randn(len(x))

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the scatter plot
ax.scatter(x, y)

# Fit a regression line
m, c = np.polyfit(x, y, 1)
ax.plot(x, m * x + c, color='red')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Trend Scatter Plot')

# Show the plot
plt.show()
```

#### Moving Average Plot:
A moving average plot is useful for smoothing out fluctuations and highlighting long-term trends in time series data. It calculates the average value over a specified window of time and plots it against the corresponding time points.

```python
# Generate time series data
time = np.arange(0, 10, 0.1)
data = np.sin(time) + np.random.randn(len(time)) * 0.2

# Calculate moving average
window = 5
moving_avg = np.convolve(data, np.ones(window) / window, mode='valid')
time_ma = time[window-1:]

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the original data
ax.plot(time, data, label='Original')

# Plot the moving average
ax.plot(time_ma, moving_avg, label='Moving Average')

# Set labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Value')
ax.set_title('Trend with Moving Average')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

#### Area Plot:
An area plot can be used to visualize cumulative data or stacked values over time. It helps identify the overall pattern and relative contribution of different components to the trend.

```python
# Generate time series data
time = np.arange(0, 10, 0.1)
data1 = np.sin(time)
data2 = np.cos(time)

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the area plot
ax.fill_between(time, data1, alpha=0.5, label='Component 1')
ax.fill_between(time, data2, alpha=0.5, label='Component 2')

# Set labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Value')
ax.set_title('Trend with Stacked Area Plot')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

#### Jittering:
* Introduce random noise to the positions of data points to spread them out and minimize overlapping.
* This technique helps to maintain the original distribution while reducing the visual clutter caused by overlapping points.

```python
# Adding Random Noise
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.random.randn(100)
y = np.random.randn(100)

# Add random noise (jitter) to the x and y coordinates
jitter = 0.1  # Adjust the jitter amount to control the spreading
x_jittered = x + np.random.uniform(-jitter, jitter, size=len(x))
y_jittered = y + np.random.uniform(-jitter, jitter, size=len(y))

# Create a scatter plot with jittered data
plt.scatter(x_jittered, y_jittered)

# Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Jittered Data Points')

# Show the plot
plt.show()
```

```python
# Jittering with Categorical Data
import numpy as np
import matplotlib.pyplot as plt

# Generate categorical data
categories = ['A', 'B', 'C', 'D']
values = np.random.choice(categories, size=100)

# Add random noise (jitter) to the categorical positions
jitter = 0.1  # Adjust the jitter amount to control the spreading
values_jittered = [v + str(np.random.uniform(-jitter, jitter))for v in values]

# Create a scatter plot with jittered categorical data
plt.scatter(range(len(values_jittered)), values_jittered)

# Set labels and title
plt.xlabel('Index')
plt.ylabel('Category')
plt.title('Jittered Categorical Data')

# Set x-axis ticks and labels
plt.xticks(range(len(categories)), categories)

# Show the plot
plt.show()

```

## Multipanel Figures

Multipanel figures, also known as subplots or composite plots, allow you to display multiple plots or visualizations within a single figure. This is useful for comparing related data, presenting different views or perspectives, or showing the results of different analyses.

#### Using Matplotlib's subplots() Function:
Matplotlib provides the subplots() function to create a figure with multiple subplots arranged in a grid.

```python
import matplotlib.pyplot as plt
# Generate data

x1 = np.linspace(0, 5, 50)
x2 = np.linspace(0, 15, 50)
x3 = np.linspace(0, 20, 50)
x4 = np.linspace(0, 25, 50)

y1 = np.sin(x1)
y2 = np.cos(x2)
y3 = np.sin(x3)
y4 = np.cos(x4)

# Create a figure with 2 rows and 2 columns of subplots
fig, axes = plt.subplots(nrows=2, ncols=2)

# Access individual subplots using indexing
axes[0, 0].plot(x1, y1)
axes[0, 1].scatter(x2, y2)
axes[1, 0].bar(x3, y3)
axes[1, 1].hist(y4, bins=10)

# Set titles and labels for subplots
axes[0, 0].set_title('Plot 1')
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('Y-axis')

# Adjust spacing between subplots
plt.tight_layout()

# Show the multipanel figure
plt.show()
```

#### Using GridSpec for More Customized Layouts:
Matplotlib's GridSpec object allows you to create more customized layouts for multipanel figures.

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Create a GridSpec object with custom layout
gs = GridSpec(2, 2)

# Create subplots using the GridSpec object
ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[0, 1])
ax3 = plt.subplot(gs[1, :])

# Plot data in subplots
ax1.plot(x1, y1)
ax2.scatter(x2, y2)
ax3.bar(x3, y3)

# Set titles and labels for subplots
ax1.set_title('Plot 1')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')

# Adjust spacing between subplots
plt.tight_layout()

# Show the multipanel figure
plt.show()
```

#### Using Seaborn for Faceted Plots:
The Seaborn library provides a convenient way to create multipanel figures with faceted plots using its FacetGrid or PairGrid functions

```python
# FacetGrid
import seaborn as sns

# Load example dataset
tips = sns.load_dataset('tips')

# Create a FacetGrid with scatterplot
g = sns.FacetGrid(data=tips, col='time', row='sex')
g.map(sns.scatterplot, 'total_bill', 'tip')

# Show the plot
plt.show()
```

```python
# PairGrid
import seaborn as sns

# Load example dataset
iris = sns.load_dataset('iris')

# Create a PairGrid with scatterplot
g = sns.PairGrid(data=iris, vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)

# Add a legend
g.add_legend()

# Show the plot
plt.show()
```
